Import the Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

 Reading data from remote url

 url="http://bit.ly/w-data"
data=pd.read_csv(url)


Visualising Dataset

```python
print("Dataset:",data.shape)
print("Dataset size:",data.size)
print("Head Elements:",data.head())
print("Tail Elements:",data.tail())
```

Creating a Dataframe

```python
df=pd.DataFrame(data)
hours=df['Hours'].head(12)
scores=df['Scores'].head(12)
```

Bar Plot of Dataset

```python
fig=plt.figure(figsize=(10,7))

#Horizontal Bar Plot
plt.bar(hours,scores)
plt.title('Marks vs Study Hours',size=20)
plt.ylabel('Marks Percentage',size=12)
plt.xlabel('Hours Studied',size=12)

#show plot
plt.show()
```

Scatter Plot for visualising Distribution of Scores

```python
fig=plt.figure(figsize=(10,7))

#Scatter Plot
data.plot(x='Hours',y='Scores',style='o')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Scored')
plt.show()
```

Plotting Regression Plot

```python
fig=plt.figure(figsize=(10,7))

#Regression Plot
sns.regplot(x=data['Hours'],y=data['Scores'])
plt.title('Regression Plot')
plt.ylabel('Percentage Scored')
plt.ylabel('Hours Studied')
plt.show()
```

Observing Correlation

```python
#Observing Correlation
print(data.corr())
```

Splitting the Dataset into Training and Test Dataset

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
from sklearn .linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

Training the Model

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
print("Training completed")
```

Predicted Score if a student studies for 9.25 hrs/day

```python
hours=[9.25]
result=regression.predict([hours])
print("Expected score if the student studies for 9.25 hrs is:{}".format(round(result[0],2)))
```

Testing the Model

```python
y_pred=reg.predict(X_test)
```

```python
actual_predicted=pd.DataFrame({'Target':y_test,'predicted':y_pred})
actual_predicted
```

Analysing the Model

```python
#Calculating Accuracy of Model

from sklearn import metrics
from sklearn.metrics import r2_score
print('Mean Absolute error:',metrics.mean_absolute_error(y_test,y_pred))
print('R2 Score:',r2_score(y_test,y_pred))
```
